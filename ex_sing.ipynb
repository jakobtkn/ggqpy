```python
import ggqpy
from sympy import Symbol, cos, plot, lambdify

from IPython.display import display
from ggqpy.functionfamiliy import FunctionFamily, Interval
```

```python
x = Symbol("x", real=True)
weight = cos(x)

F = FunctionFamily(Interval(-1,1), [weight*x**k for k in range(40)])
```

```python
fsym = (0.4*x**3 + 0.3*x**4)*cos(x)
integral = float(F.target_integral(fsym))
display(fsym)
plot(fsym,xlim=(-1,1),ylim=(-0.1,10))
print(integral)
```

```python
x,w = ggqpy.generalized_gaussian_quadrature(F)
```

```python
f = lambdify('x',fsym,"numpy")
Iquad = f(x)@w
print(Iquad-integral)
```
