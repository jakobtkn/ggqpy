```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
n = 10
xx = np.linspace(-1,1,100)
f = lambda x: -0.5 + x**2
g = lambda x: 0.5 + -x**2
x, w = np.polynomial.legendre.leggauss(n)
plt.plot(xx,f(xx))
plt.plot(xx,g(xx))
print(x)
print(w)
print((f(x)*w).sum())
print(quad(f,-1,1))
print((g(x)*w).sum())
print(quad(g,-1,1))
```
